```python
import requests
```

```python
url = "https://raw.githubusercontent.com/Audrine-m/trade_data/refs/heads/main/MOCK_DATA.json"
```

```python
r = requests.get(url)
```

```python
r.status_code
```

```python
MOCK_DATA = r.json()
print(type(MOCK_DATA))
```

```python
Trading_data = MOCK_DATA["Trading_data"]
```

```python
print(type(Trading_data))
```

```python
product_data = requests.get(product_url)
```

```python
product_data.status_code
```

```python
dummy_data = product_data.json()
```

```python
print(type(dummy_data))
```

```python
products = dummy_data["products"]
```

```python
carts_url = "https://dummyjson.com/carts"
carts_data = requests.get (carts_url)
data = carts_data.json()
print(type(data))
carts = data["carts"]
```

```python
import pandas as pd
```

```python
import pandas as pd

df = pd.DataFrame(Trading_data)

# Remove any accidental index column
if 'index' in df.columns:
    df = df.drop(columns=['index'])

# Reset dataframe index
df.reset_index(drop=True, inplace=True)

# Save without index
df.to_csv("trading_data.csv", index=False)
df
```
